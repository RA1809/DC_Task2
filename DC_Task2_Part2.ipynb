```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv(r'data.csv')
#df.head()
#print(df.dtypes)
```

```python
Age=df['Age']
mi=df['Monthly Income']
result = np.vstack((Age, mi)).T
sortedArr = result[result[:,0].argsort()]
plt.plot(sortedArr[:,0],sortedArr[:,1])
```

thus we can see that after a certain age, as age increases salary increases(maybe since age can be considered to be equivalent to experience and hence expertise int the field)

```python
Age=df['Age']
des=df['Designation']
result = np.vstack((Age,des)).T
sortedArr = result[result[:,0].argsort()]
plt.plot(sortedArr[:,0],sortedArr[:,1])
```

As age increases, the position of the person in the job hierarachy goes up.

```python
df['Engagement Score (% Satisfaction)']=df['Engagement Score (% Satisfaction)'].astype(str)
#print(df['Engagement Score (% Satisfaction)'])
df['Engagement Score (% Satisfaction)'] = df['Engagement Score (% Satisfaction)'].apply(lambda x: float(x.split()[0].replace('%', '')))
#print(df['Engagement Score (% Satisfaction)'])
a=df['Engagement Score (% Satisfaction)']
df['Engagement Score (% Satisfaction)'] = np.asarray(a, dtype=float)
#print(df['Engagement Score (% Satisfaction)'])
#print(df.dtypes)
```

```python
import seaborn as sns
sns.barplot(x='Zone', y='Engagement Score (% Satisfaction)', data=df);
```

We can see average of engagement score is more or less same across all regions.
